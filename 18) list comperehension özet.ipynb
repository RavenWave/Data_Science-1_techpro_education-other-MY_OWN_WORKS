# LIST COMPREHENSION KONUSU İLE İLGİLİ ÖNEMLİ GÖRDÜKLERİM

    Aslında List Comprehension'lar, (list + for + if) fonksiyonlarının birlikte ve tek bir satırda kullanılmasına olanak sağlayan pratik bir yöntemdir

    Kullanım Alanları aşağıdakilerle sınırlıdır:
        1) for kullanımı
        2) for + if kullanımı
        3) for + if + else kullanımı
        3) for + for (iç içe for)

> NOT: break ve continue'ler list comprehension içerisinde kullanılamaz

### Sırayla Örnekler:

```python
[print(i) for i in 'techpro']
```

```python
[print(i, '-', i * 2   ) for i in 'techpro']
```

```python
liste_1 = [1, 2, 3, 4, 5]
[print(i ** 2) for i in liste_1]
```

```python
liste_1 = [1, 2, 3, 4, 5]
[i ** 2 for i in liste_1]
```

```python
[print(f'{i} sayısının karesi {i ** 2}', end=' | ') for i in (1, 2, 3, 4, 5)]
```

```python
isimler = ['EmrE', 'iREm', 'süheyla']

[i.capitalize() for i in isimler]
```

```python
[print(i, end='-') for i in 'techpro']
```

```python
dict_1 = {'name': "dilek", "age": 26, "job": "data_scientist"}

[i for i in dict_1]
```

```python
a = [i for i in dict_1.items()]
print(a)
print()
print(a[2])
```

```python
[print(i, j, sep='|') for i, j in dict_1.items()]
```

```python
[print(f'key: {i:5}| value: {j}') for i, j in dict_1.items()]
```

```python
# tek sayılar listesi:

[i for i in range(1, 50, 2)]
```

```python
# çift sayılar lsitesi:

[i for i in range(2, 51, 2)]
```

    Ödev: Kullanıcıdan 1 ve 9 arasında bir sayı isteyin (1 ve 9 dahil). Bu sayının çarpım tablosunu oluşturun.

    ÖR: input = 8
    8 x 0 = 0
    8 x 1 = 8 
    ...
    8 x 10 = 80

```python
x = int(input('bir sayı gir. (1-9 arasında)'))

[print(f'{x} x {i} = {x * i}') for i in range(1, 10)]
```

```python
x = sum([i for i in range(50)])

print(x)
```

    1'den 100'e kadar olan çift sayıların toplamını bul 

```python
x = sum([i for i in range(2, 1001, 2)])
print(f"1'den 1000'e kadar olan çift sayıların toplamı: {x}")
```

    250'den 2500'e kadar olan tek sayıların toplamını bul

```python
x = sum(i for i in range(251, 2500))
print(f"250'den 2500'e kadar olan tek sayıların toplamı: {x}")
```

```python
sayılar = [12, 252, 11, 63, 67, 23, 654, 4, 125]

[i for i in sayılar if i >= 12]
```

```python
x = int(input('sayı giriniz. '))

[print(f'{x} x {i} = {x * i}') for i in range(1, 10) if x < 10 and x > 0]
```

```python
isim_soyisim = input('isim soyisim: ')

[print(i) for i in isim_soyisim if 'm' in isim_soyisim]
```

```python
sayılar_listesi = [1, 2, 3, 4, 5]

# yukarıdaki listenin içerisindeki elemanların karesini yazdırın. 3'ü dahil etmesin.

[i ** 2 for i in sayılar_listesi if i != 3]
```

    aşağıdaki listede string olan elemanların uzunluklarını listeleyelim

```python
karışık = ['emre', 23, 'yirmiüç', 'irem' 'emre çakır', 12, 'irem süheyla çukurlu']

[len(i) for i in karışık if isinstance(i, str)]
```

    aşağıdaki listede integer olan elemanların 25. kuvvetlerini listeleyelim

```python
karışık = ['emre', 23, 'yirmiüç', 'irem' 'emre çakır', 12, 'irem süheyla çukurlu']

[i ** 25 for i in karışık if isinstance(i, int)]
```

![list comprehensiyon_ 3. Durum.jpg](<attachment:list comprehensiyon_ 3. Durum.jpg>)

```python
liste = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# yukarıdaki listede tek olan sayıların karesini, çift olan sayıların ise iki katını alıp bir listeye koy

[i ** 2 if i % 2 else i * 2 for i in liste]
```

```python
mix = [234, 'emre çakır', 'süheyla', 12, 2, 0, 'emre', 'süheyla çukurlu', -3]

# yukarıdaki listeyi ingeterların ve aynı zamanda len'i 10'dan büyük olan stringlerin olduğu listeye çevir. şarta uymayanlar için 'x: eleman' yazsın

x = [i if (isinstance(i, int) and i > 10) or (isinstance(i, str) and len(i) > 10) else f"x: {i}" for i in mix]

print(x)
```
