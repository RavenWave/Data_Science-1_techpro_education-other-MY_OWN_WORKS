# FOR DÖNGÜSÜ
#### Başlangıç, bitiş ve artış değerleri belirtilen işlemleri tek tek tekrar eder.

```python

print(">> Aşagidaki sayilar 1'den 100'e kadar olan tüm sayilar:\n")
for i in range(1,100):
    print(i, end=" ")
    
print("\n\n>> Aşagidaki sayilar 1'den 100'e kadar olan tüm tek sayilar:\n")    
for i in range(1,100):
    if i % 2 != 0 :
        print(i, end=" ") 
        
print("\n\n>> Aşagidaki sayilar 1'den 100'e kadar olan tüm tek sayilar:\n")    
for i in range(1,100,2):
    print(i, end=" ")    
    
print("\n\n>> 21'den 0'a kadar olan sayilari 3'er 3'er azaltarak yazdiralim.:\n")    
for i in range(21,0,-3):
    print(i,end=" ")
    
print("\n\n>> 1'den 100'e kadar olan sayilarin toplami:\n")
i = 0
for n in range(1,100):
    i += n
print(i)

print("\n>> 10 ve 10'un üstlerini yazdiralim, 10^5 ye kadar yazdiralim:\n")
i = 10
for n in range(1,6):
    a = i**n 
    print(a,end=" -- ")

print("\n\n>> 10 ve 10'un üstlerini yazdiralim, 10^5 ye kadar yazdiralim:\n")
for i in range(1,6):
    print("{}".format(10**i),end=" -- ")
```

# While Döngüsü
#### Belirlenen koşul sağlanmaya devam ettiği sürece devam eden bir dönügüdür. 

```python
x = 1
while x < 10:
    print("Kod yazmak güzel şey!", end=" --> ")
    x += 2
```

```python
# 1'den 16'ya kadar olan çift sayıları yazdırma:

x = 0
while x < 17:
    print(x,end=" - ")
    x += 2
```

# "break" komutu döngüyü sonlandırır.

```python
# Eğer aşağıda break komutu kullanılmasaydı bu döngü sonsuz devam edecekti.

while True:
    print("Emre Çakir kod yaziyor..")
    break 
```

```python
while True:
    sayi1 = int(input("Birinci sayiyi giriniz (çikiş yapmak için '0' harfine basmaniz yeter): "))
    sayi2 = int(input("İkinci sayiyi giriniz (çikiş yapmak için '0' harfine basmaniz yeter): "))
    if sayi1 == 0 and sayi2 == 0:
        break
    else:
        print(sayi1 + sayi2)
```

```python
# Kullanıcıya bir parola belirletirken parolanın 6 karakterden kısa olmasını sağlayalm.

while True:
    parola = input("Parola Belirleyin: ")
    parola_uzunluk = len(parola)
    if parola_uzunluk > 6:
        print("Belirlediğiniz parola",parola_uzunluk,"karakterden oluşuyor. Lütfen en fazla 6 karakterli bir şifre belirleyiniz.")
    else:
        print("Parolaniz basariyle olusturuldu!")
        break
```
