# HATA ÇEŞİTLERİ


### **_1. Söz Dizimi Hatası (SyntaxError)_**
#### a) Hatalı atama işlemi
#### b) Eşleşmeyen parantez
#### c) Eşleşmeyen tırnak işareti
#### d) Hatalı girinti (if, while gibi döngülerde girintiler önemlidir)
### **_2. Çalışma Zamanı Hataları_**
#### a) Atanamyan değişken
#### b) Sıfıra bölme işlemi
#### c) Sayıları string'e bölme
#### d) Gerçekleşmesi imkansız olan koşullar
### **_3. Mantık Hataları (Anlam Bilimsel Hatalar)_**
#### a) Program genellikle hata mesajı vermeden çalışır ancak istenilen işlemi gerçekleştirmez. En zor hata ayıklama türüdür.
